```python
import pandas as pd
import csv
import os
```

```python
fileDir = os.path.abspath(os.path.dirname(""))
usa_csv = os.path.join(fileDir, "name_usa.csv")
kor_csv = os.path.join(fileDir, "name_kor.csv")
db_usa= pd.read_csv(usa_csv)
db_kor= pd.read_csv(kor_csv)
```

```python
db_kor.tail()
# db_usa.tail()
```

```python
usa_last_name = [name for name in db_usa["last_name"] if str(name) != 'nan']
usa_first_name = [name for name in db_usa["first_name"] if str(name) != 'nan']
kor_last_name = [name for name in db_kor["last_name"] if str(name) != 'nan']
kor_first_name = [name for name in db_kor["first_name_com"] if str(name) != 'nan']
kor_first_name
```

```python
#Combining first and last names
usa_full_name = []
kor_full_name = []

for l_name in usa_last_name: 
    for f_name in usa_first_name:
       usa_full_name.append(l_name.lower()+ " " +f_name.lower())
    
for l_name in kor_last_name: 
    for f_name in kor_first_name:
       kor_full_name.append(l_name.lower()+ " " +f_name.lower())
```

```python
df_output_usa = pd.DataFrame({'name':usa_full_name})
df_output_kor = pd.DataFrame({'name':kor_full_name})
print(df_output_usa)
print(df_output_kor)
```

```python
df_output_usa.to_csv("name_usa_combined_space.csv")
df_output_kor.to_csv("name_kor_combined_space.csv")
```
